```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

f= open("../AMLD_2020.ipynb","r+")
contents =f.read().lower()

words_of_interest = ['OSRM', 'OSM', 'distance', 'euclidean', 'manhattan', 'haversine', 'route', 'duration', 
                     'KNNRegressor', 'H3', 'latitude', 'longitude', 'projection', 'UTM', 'geometry', 'pipeline',
                     'fastest', 'shortest', 'R-Tree', 'way', 'relation', 'GPS', 'map matching', 'kepler',
                     'osmnx', 'segment']
words_of_interest = [elem.lower() for elem in words_of_interest]

words_of_interest_f = [elem for elem in contents.split(' ') if elem in words_of_interest]
words_of_interest_f = ' '.join(words_of_interest_f)
words_of_interest_f

wordcloud = WordCloud(max_words=100, background_color="white").generate(words_of_interest_f)
plt.figure(figsize=[18,5])
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
